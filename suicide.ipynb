```python
mortality = pd.read_csv("/Users/jiahaoshen/Desktop/NCHS_-_Leading_Causes_of_Death__United_States.csv", header=0)
mortality.head()
```

```python
suicide = mortality[mortality['Cause Name'] == 'Suicide']
```

```python
suicide = suicide[suicide['State']!= 'District of Columbia']
```

```python
suicide.columns
```

```python
suicide.drop(columns=['113 Cause Name','Cause Name','Age-adjusted Death Rate'])
```

```python
suicide_year_state = suicide.groupby(['Year', 'State']).sum()
```

```python
suicide_year_state['Deaths'].sum()
```

```python
suicide_year = suicide.groupby(['Year']).sum()
suicide_year
```

```python
fig, ax = plt.subplots(figsize=(15,11))
suicide_year_state['Deaths'].unstack().plot(ax=ax, xticks = suicide_year_state.reset_index()['Year'])
ax.set_xlabel('Year')
ax.set_ylabel('Death')
ax.set_ylim([0, 450000])
```

```python
fig, axes = plt.subplots(1, 1, figsize=(12, 6), dpi=100)
line1, = axes.plot(suicide_year.index, suicide_year['Deaths'], color='indianred', linewidth=3)


axes.set_xticks(suicide_year.index)
for tick in axes.xaxis.get_major_ticks():
    tick.label.set_fontsize(16) 
plt.yticks(fontsize=18)
plt.xticks(rotation=45)
plt.grid(b=True)
plt.show()
```
