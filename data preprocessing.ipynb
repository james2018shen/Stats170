# Unemployment

```python
employ1 = pd.read_csv("data/CAEMP25S__ALL_STATES_1969_2000.csv", encoding = "Latin-1")
```

```python
employ2 = pd.read_csv("data/CAEMP25N__ALL_STATES_2001_2017.csv",encoding = "Latin-1")
```

```python
employ1 = employ1[employ1['Description']=='Total employment (number of jobs)']
employ1=employ1[['GeoName','Description','1999','2000']]
employ1 = employ1[~employ1.GeoName.str.contains(",")]
employ1 = employ1[1:52]
employ1 = employ1.drop(8096)
```

```python
employ1.reset_index(drop=True, inplace=True)
employ1
```

```python
employ2 = employ2[employ2['Description']=='Total employment (number of jobs)']
employ2=employ2[['GeoName','Description','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011',
                '2012','2013','2014','2015','2016']]
employ2 = employ2[~employ2.GeoName.str.contains(",")]
employ2 = employ2[1:52]
employ2 = employ2.drop(11616)
```

```python
employ2.reset_index(drop=True, inplace=True)
#employ2 = employ2.drop(['index'],axis=1)
employ2
```

```python
employ2 = employ2.drop(columns=['GeoName','Description'])
```

```python
employ = pd.concat([employ1, employ2], axis=1)
```

```python
employ
```

```python
employ=employ.sort_values(by=['GeoName'])
employ = employ.as_matrix(columns=['1999', '2000', '2001', '2002', '2003', '2004', '2005',
       '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014',
       '2015', '2016'])
employ = employ.astype('int64')
employ.shape
```

```python
np.save("/Users/jiahaoshen/Desktop/Stats170/newdata/employ.npy",employ)
```
