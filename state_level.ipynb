### How to transfer to STATE | YEAR | GDP

```python
income = pd.read_csv("data/SAINC5N_1998_2018_ALL_AREAS.csv")
income.head()
```

```python
income = income[(income["Description"]=='Personal income (thousands of dollars)')
       |(income["Description"]=='Population (persons) 2/')
       |(income["Description"]=='Per capita personal income (dollars) 3/')]

```

```python
income.columns
```

```python
income = income.drop(columns=['GeoFIPS','Region','TableName','Unit','IndustryClassification','IndustryClassification','LineCode'])
income.head()
```

```python
income.rename(columns={'GeoName': 'State'}, inplace=True)
```

```python
income = income.reset_index(drop=True)
```

```python
income = income[(income["State"]!='New England')
       &(income["State"]!='Mideast')
       &(income["State"]!='Great Lakes')
       &(income["State"]!='Plains')
       &(income["State"]!='Southeast')
       &(income["State"]!='Southwest')
       &(income["State"]!='Rocky Mountain') 
       &(income["State"]!='Far West')
       &(income["State"]!='District of Columbia')]
```

```python
pce = pd.read_csv("data/SAEXP1_1997_2017_ALL_AREAS_.csv")
pce.head()
```

```python
pce = pce[pce["Description"]=='Personal consumption expenditures']
pce = pce.drop(columns=['GeoFIPS','Region','TableName','ComponentName','Unit','Line','IndustryClassification'])
pce.head()
```

```python
pcep = pd.read_csv("data/SAEXP2_1997_2017_ALL_AREAS_.csv")
pcep.head()
```

```python
pcep = pcep[pcep["Description"]=='Per capita personal consumption expenditures']
pcep = pcep.drop(columns=['GeoFIPS','Region','TableName','ComponentName','Unit','Line','IndustryClassification'])
pcep.head()
```

```python
pce.rename(columns={'GeoName': 'State'}, inplace=True)
pcep.rename(columns={'GeoName': 'State'}, inplace=True)
pce = pce[pce["State"]!='District of Columbia']
pcep = pcep[pcep["State"]!='District of Columbia']
pce = pce.reset_index(drop=True)
pcep = pcep.reset_index(drop=True)
```

# COMBINE

```python
gg = pd.concat([pce, pcep,income,gdp], sort=False)
gg
```

```python
gg.sort_values(['State'], inplace=True)
```

```python
gg
```
